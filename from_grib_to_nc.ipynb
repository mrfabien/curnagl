```python
# load multiple files and combine them

year = 2021
var = 'convective_available_potential_energy'
short_var = 'cape'
number_var = '059'

ds = xr.open_mfdataset('/work/FAC/FGSE/IDYST/tbeucler/default/raw_data/ECMWF/ERA5/grib/e5.oper.an.sfc.128_'+number_var+'_'+short_var+'.ll025sc.'+str(year)+'*.grb', engine='cfgrib')
```

```python
# extract the data with only a 3 hour interval

ds = ds.sel(time=ds.time.dt.hour.isin([0,3,6,9,12,15,18,21]))
ds

# remove numer, step, surface and valid_time coordinates

ds = ds.drop(['number','step','surface','valid_time'])
ds
```

```python
# convert the file to netcdf with a timestep of 3 hours

ds.to_netcdf('/work/FAC/FGSE/IDYST/tbeucler/default/raw_data/ECMWF/WS_fabien/missing_var/Temp_ERA5_'+str(year)+'_'+var+'.nc')
```
