```python
import os
import re

# Define the directory containing the files
directory = '/scratch/faugsbur/relative_humidity'

# Define the variable you want to add
variable = 'relative_humidity'  # Change this to the actual variable you want to include

# Define the regex pattern to match the filenames
pattern = re.compile(r'ERA5_(\d{4})-(\d{1,2})–relative_humdity_(\d+).nc')

# ERA5_1992-10–relative_humdity_relative_humidity30
# ERA5_1992-12–relative_humdity_20

# Iterate over all files in the directory
for filename in os.listdir(directory):
    # Check if the filename matches the pattern
    match = pattern.match(filename)
    if match:
        year, month, level = match.groups()
        new_filename = f'ERA5_{year}-{month}_{variable}_{level}.nc'
        # Get full paths
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_filename)
        # Rename the file
        os.rename(old_file, new_file)
        print(f'Renamed: {old_file} -> {new_file}')

print('Renaming complete.')
```
