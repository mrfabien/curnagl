```python
import os
import csv
import pandas as pd

def check_folder_for_files(folder_path, variable_names_csv):
    # Load variable names from CSV
    variable_names = []
    with open(variable_names_csv, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        for row in csvreader:
            variable_names.extend(row)

    # Initialize a dictionary to store expected file counts for each year-month-variable combination
    expected_counts = {}
    for year in range(1990, 2022):
        for month in range(1, 13):
            for variable_name in variable_names:
                expected_counts[f"{year:04d}-{month:02d}_{variable_name}.nc"] = 0

    # Count actual files in the folder
    actual_counts = {}
    for filename in os.listdir(folder_path):
        if filename.endswith('.nc') and filename.startswith('ERA5'):
            actual_counts[filename] = actual_counts.get(filename, 0) + 1

    # Check if the actual counts match the expected counts
    for expected_filename, expected_count in expected_counts.items():
        if actual_counts.get(expected_filename, 0) != expected_count:
            print(f"Expected {expected_count} files named {expected_filename}, but found {actual_counts.get(expected_filename, 0)}.")
```

```python
import os
import pandas as pd

def check_folders_for_files(root_folder, variable_names_csv):
    # Load variable names from CSV using pandas
    variable_names_df = pd.read_csv(variable_names_csv, header=None)
    variable_names = variable_names_df[0].tolist()

    # Initialize a dictionary to store expected file counts for each year-month-variable combination
    expected_counts = {}
    for year in range(1990, 2022):
        for month in range(1, 13):
            for variable_name in variable_names:
                expected_counts[f"{year:04d}-{month:02d}_{variable_name}.nc"] = 0

    # Iterate over each variable folder
    for variable_name in variable_names:
        folder_path = os.path.join(root_folder, variable_name)
        if not os.path.exists(folder_path):
            print(f"Folder {folder_path} not found.")
            continue

        # Count actual files in the folder
        actual_counts = {}
        for filename in os.listdir(folder_path):
            if filename.endswith('.nc') and filename.startswith('ERA5_'):
                actual_counts[filename] = actual_counts.get(filename, 0) + 1

        # Check if the actual counts match the expected counts
        for expected_filename, expected_count in expected_counts.items():
            if actual_counts.get(expected_filename, 0) != expected_count:
                print(f"Expected {expected_count} files named {expected_filename} in {variable_name} folder, but found {actual_counts.get(expected_filename, 0)}.")

# Example usage:
root_folder = '/work/FAC/FGSE/IDYST/tbeucler/default/fabien/repos/curnagl/missing_var/'
variable_names_csv = 'missing_variables_test.csv'
check_folders_for_files(root_folder, variable_names_csv)
```
